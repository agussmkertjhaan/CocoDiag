```python
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
import json
```

Memuat model

```python
model = load_model('model/image_classification_model.h5')
```

Memuat class indices

```python
with open('model/class_indices.json', 'r') as f:
    class_indices = json.load(f)
class_labels = list(class_indices.keys())
```

Fungsi untuk memprediksi gambar baru dengan informasi tambahan

```python
def predict_image(image_path):
    img = image.load_img(image_path, target_size=(150, 150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    predictions = model.predict(img_array)[0]  # Mengambil hasil prediksi untuk gambar tunggal
    top_indices = predictions.argsort()[-3:][::-1]  # Mengambil 3 indeks teratas dan membalik urutannya
    top_labels = [class_labels[i] for i in top_indices]
    top_scores = [predictions[i] for i in top_indices]
    print("Top 3 Predictions:")
    for label, score in zip(top_labels, top_scores):
        print(f"Class: {label}, Score: {score:.2f}")

    # Menampilkan probabilitas untuk semua kelas
    print("\nProbabilities for all classes:")
    for i, prob in enumerate(predictions):
        print(f"{class_labels[i]}: {prob:.2f}")
```

Contoh penggunaan fungsi prediksi

```python
predict_image('asd1.jpg')  # Ganti dengan path ke gambar baru
```
